```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define position and dimension model
position = np.arange(100)  # 100 positions in the sequence
d_model = 512  # Typical size of the model dimension
```

```python
# Compute sinusoidal functions for positional encoding
angles = position[:, np.newaxis] / np.power(10000, (2 * np.arange(d_model//2)) / d_model)
sin_angles = np.sin(angles)
cos_angles = np.cos(angles)
```

```python
# Plotting the first few dimensions of the positional encoding
plt.figure(figsize=(10, 6))
plt.plot(position, sin_angles[:, 0], label='sin (1st dim)')
plt.plot(position, cos_angles[:, 0], label='cos (1st dim)')
plt.plot(position, sin_angles[:, 1], label='sin (2nd dim)')
plt.plot(position, cos_angles[:, 1], label='cos (2nd dim)')
plt.title('sin + cosine Functions of Positional Encoding')
plt.xlabel('Position')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```
