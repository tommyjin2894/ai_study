# 역전파 알고리즘의 문제점
- 경사 소실 (gradient vanishing)
  그레디언트 값이 매우 작아져 학습이 안되는 현상
  0 에 가까워 지면 가중치 업데이트가 안된다



### 데드 렐루 문제



미분값이 0이어도 학습에 효과가 있다는 연구 결과 <br>
양수 값만 있어도, 오히려 학습에 도움이 된다. <br>

### 과적합
Generalization > Regularization
 
- Regularization 
: 가중치 정규화 - 모델이 학습 데이터에 너무 학습을 해서 새 데이터에 대한 성능이 떨어지는 현상 방지
    모델 관점 :
        1. L1 규제 (라쏘) : 가중치를 0으로 만들어 모델을 단순화 (드롭 아웃과 비슷)
        2. L2 규제 (릿지) : 가중치의 값을 줄여 모델의 복잡도를 낮춤
        3. 드롭 아웃 : 무작위 노드 비활성 (Test 시에는 모든 노드 사용)
        4. 조기 종료 : 성능이 개선 되지 않을 시에는 조기 종료하고 이전 학습 내용을 복구한다.
    데이터의 관점의 : 
        5. 데이터 확장 (Augmentation) : 데이터를 다양화(이미지 줄이고 늘리고 자르고)
        6. 교차 검증 (cv)
