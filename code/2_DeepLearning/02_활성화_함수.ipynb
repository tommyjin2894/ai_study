### 활성화 함수

```python
import numpy as np
import matplotlib.pyplot as plt

# 시그모이드(Sigmoid) 활성화 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 하이퍼볼릭 탄젠트(Hyperbolic Tangent) 활성화 함수 정의
def tanh(x):
    return np.tanh(x)

# 렐루(ReLU) 활성화 함수 정의
def relu(x):
    return np.maximum(0, x)

# 리키 렐루(Leaky ReLU) 활성화 함수 정의
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# 엘루(ELU) 활성화 함수 정의
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# 소프트플러스(Softplus) 활성화 함수 정의
def softplus(x):
    return np.log(1 + np.exp(x))

# GeLU(Gaussian Error Linear Unit) 활성화 함수 정의
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

# 예시 데이터 생성
x = np.linspace(-10, 10, 100)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_elu = elu(x)
y_softplus = softplus(x)
y_gelu = gelu(x)

# 활성화 함수의 결과를 시각화
def graph(func,name):
    _, ax = plt.subplots(figsize = (7,5))
    ax.plot(x, func)
    ax.set_title(f'{name}')
```


### **Sigmoid (시그모이드) 함수**:
$$\phi = \frac{1}{1+e^{-x}}$$
   - 범위: (0, 1)
   - 비선형성을 부여하여, 이진 분류 문제에 유리.
   - 입력 값이 크거나 작을 경우 경사하강이 거의 일어나지 않는 **'Vanishing Gradient'** 문제 발생 가능성이 큼


```python
graph(y_sigmoid,'Sigmoid')
```


### **Tanh (하이퍼볼릭 탄젠트) 함수**:
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$
   - 범위: (-1, 1)
   - 시그모이드와 비슷하지만, 출력 값의 중심이 0에 더 가깝기 때문에 시그모이드 대비 수렴 속도가 빠른 편.
   - 역시 'Vanishing Gradient' 문제를 가질 수 있지만, 시그모이드보다 덜 심각.


```python
graph(y_tanh,'tanh')
```


### **ReLU (Rectified Linear Unit) 함수**:
$$f(z) = max(0, z)$$
   - 범위: [0, ∞)
   - 간단한 구현과 빠른 수렴 속도를 제공하여 많이 사용됩니다.
   - 입력 값이 0 이하일 때 출력을 0으로 처리하여 **'Dying ReLU'** 문제를 일으킬 수 있음.


```python
graph(y_relu,'ReLU')
```


### **Leaky ReLU 함수**:
$$f(z) = max(\epsilon z, z)$$
   - ReLU 문제를 개선하여, 활성화 값이 0 이하일 때 작은 기울기를 유지하게끔 수정한 함수.
   - 매개변수 \(\epsilon\) 값에 따라 비활성화 영역에서 작은 음의 기울기를 허용.


```python
graph(y_leaky_relu,'Leaky ReLU')
```

### **ELU (Exponential Linear Unit) 함수**:
$$f(x) = \begin{cases}
    x & \text{if } x \geq 0 \\
    \alpha (e^x - 1) & \text{if } x < 0
    \end{cases}$$
   - 활성화 값이 0 이상일 때 ReLU와 동일하나, 0 미만일 때는 지수 함수의 형태를 가진다.
   - 음수 입력 값에 대해 기울기를 유지하여 'Dying ReLU' 문제를 완화하고, 출력의 평균을 0 근처로 유지하여 수렴 속도를 향상.


```python
graph(y_elu,'ELU')
```

### **SoftPlus 함수**:
$$f(z) =  \ln(1 + e^x)$$
   - ReLU의 부드러운 근사치로 볼 수 있으며, 항상 미분 가능한 특성을 가짐.
   - 입력 값에 대한 경사를 항상 제공하여 모든 입력 값에 대해 학습이 가능 하게 함 .


```python
graph(y_softplus,'SoftPlus')
```

### **GeLU (Gaussian Error Linear Unit) 함수**:
$$ f(x) = 0.5 \times x \times \left( 1 + \tanh\left(\sqrt{\frac{2}{\pi}} \times \left(x + 0.044715 \times x^3 \right)\right) \right) $$
   - 자연어 처리 분야의 Neural Networks에 주로 사용되며, 비선형 변환을 가우시안 확률 기반으로 수행.
   - ReLU와 비슷하지만, 더 높은 성능을 보이는 경우가 흔함.
   - 복잡도가 조금 더 있지만, 최신 모델에서 효과적이라고 입증됨.

```python
graph(y_gelu,'GeLU')
```
