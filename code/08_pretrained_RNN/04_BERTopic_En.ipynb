### Bert (Bidirectional Encoder Representations from Transformers)
- Pre-trained Language Mdodel
- 특징 : 양방향 학습, fine-tuning 이 용이하다
- Transformer 의 Encoder 블록만 이용

- MLM(masked language model)
    - 입력 텍스트의 일부 단어를 무작위로 마스킹 후 문맥으로 예측
    - 왼쪽 과 오른쪽 문맥을 동시에 (전체를 동시에) 고려
    
- Transformer
    - self-attention
    - 병렬 처리로 모든 단어 동시 연관성 계산
    
- Pre-trained, Fine-tunning
    - 대규모 corpus로 사전 훈련 → 일반화가 잘되어있음
    - 감정, 질의 응답에 대해 fine-tunning → 성능이 좋다
    
- 토큰<br>
    ## 예시 : [CLS] i'm good [SEP] Thanks.
    - **[CLS]** : 입력 시퀀스의 전체 정보 반영 (분류 작업)
    - **[SEP]** : 문장 사이의 구분 (ex. 질의 응답, 자연어 추론 등)
    - Token, Segment, Positional Embeddings
        - Token : 문장의 각 단어의 벡터표현
        - Segment : 각 문장이 어느 문장에 속해 있는지
        - Positional Embedding : 위치 순서 정보
        
        | Input | [CLS] | i’m | good | [SEP] | Thanks. |
        | --- | --- | --- | --- | --- | --- |
        | Token | $\text{E}_{\text{[CLS]}}$ | $\text{E}_{\text{i’m}}$ | $\text{E}_{\text{good}}$ | $\text{E}_{\text{[SEP]}}$ | $\text{E}_{\text{Thanks.}}$ |
        | Segment | $\text{E}_a$ | $\text{E}_a$ | $\text{E}_a$ | $\text{E}_a$ | $\text{E}_b$ |
        | Positional Embeddings | $\text{PE}_0$ | $\text{PE}_1$ | $\text{PE}_2$ | $\text{PE}_3$ | $\text{PE}_4$ |
    - 특이 사항
        - hidden state vector 를 embedding vector
            
            Bert : 768 dim
            
            Transformer : 512 dim
        

- 하이퍼 파라미터 및 파라미터

    |  | BERT Base | BERT Large |
    | --- | --- | --- |
    | 층 | 24 | 112 |
    | Hidden Size
    (Embedding) | 768 | 1024 |
    | Head(multi head) | 12 | 16 |
    | Parameters | 110M | 340M |

- 훈련 및 추론
    - MLM (Masked Language Model)
        - 마스킹 이용 훈련 (입력 데이터에서 15 % 랜덤 선택)
            1. 80 % [MASK] 처리
            2. 10 % 임의의 단어로 대체 (finetuning 을 위하여), 다양한 문맥을 파악
            3. 10 % 원래의 단어로 대체
    - NSP (Next Sentence Prediction)
        - 입력된 두 개의 문장이 연속된 문장인지 아닌지 예측
            - input : 두 개의 문장
            - output :  True or False (연관성이 있는지 없는지)
    - **[CLS] 의 역할**
        - 전체 문장 쌍의 문맥 정보가 들어간다

### **BERTopic**
<br>
텍스트 데이터에서 토픽 추출, 시각화하는 Python 라이브러리<br>
Transformer 모델 기반<br>
대량의 문서에서 토픽을 자동으로 추출, 토픽들 사이의 관계 파악
<br><br>
**주요 기능**<br>
자동 토픽 수 검출<br>
토픽 감소: 유사한 토픽을 합치거나 너무 작은 토픽을 제거, 명확한 토픽 구조 세팅<br>
시각화
토픽 표현: 각 토픽을 대표하는 가장 중요한 단어들을 추출하여 토픽 설명<br>
동적 토픽 모델링: 시간에 따라 변하는 토픽의 트렌드를 추적

```python
# BERTopic 클래스 임포트하기
from bertopic import BERTopic

import pandas as pd
import numpy as np
```

```python
# 데이터 읽어오기
df = pd.read_csv('df1.csv', engine='python')
df_1 = pd.read_csv('df2.csv', engine='python')

# tokyo_2020_tweets.csv : 올림픽 관련 트윗 데이터
```

```python
len(df)
```

```python
docs = df['content'].to_list() # df에서 'text' 칼럼 추출, 리스트 형태로 변환
docs_2 = df_1['content'].to_list()
docs.extend(docs_2)
```

```python
docs
```

```python
import re
```

```python
docs = [re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 .]', '', s) for s in docs]
docs = [re.sub(r'\s+', ' ', s) for s in docs]

```

```python
docs[-3:]  # 처음 3개의 문서 확인
```

```python
len(docs)
```

```python
# 모델 생성
# BERTopic 클래스의 인스턴스 생성
# language='english' 영어 데이터 처리
# nr_topics='auto'토픽 수 자동 결정  cf) nr_topics=10
# calculate_probabilities=True 각 문서가 각 토픽에 속할 확률을 계산하도록 설정합니다.

model = BERTopic(language='korean', nr_topics=10, calculate_probabilities=True)
```

```python
# 모델 설정 내용, 하이퍼파라메터 확인
# print(vars(model))
print(model)
```

```python
topics, probabilities = model.fit_transform(docs)
# 모델링 (input : 트위터 text)
# 각 문서의 topic 할당(topics)
# 각 문서가 각 topic에 속할 확률(probabilities)  리턴.
```

```python
topics  # 문서의 topic 할당 결과
```

```python
# 첫 번째 문서가 할당된  topic
topics[0]
```

```python
# topic 수
len(probabilities[0])
```

```python
probabilities[0]  # 첫 번째 문서의 각 토픽별 할당 확률
```

```python
# 가장 높은 확률 값을 가진 topic 번호
max_topic_index = np.argmax(probabilities[1])

print("topic index:", max_topic_index)
print("probabilities:", probabilities[1][max_topic_index])
```

```python
model.get_topic_info() # topic 정보 요약

# Topic: topic 고유 번호
# Count: 해당 topic에 속하는 문서 수
# Name: topic명 - topic을 대표하는 주요 단어로 단어로 구성
# Representation: topic을 가장 잘 대표하는 단어들의 리스트
# Representative_Docs: 해당 topic에 속하는 대표적인 문서들.
```

```python
model.get_topic(3)  # topic 3의 상세 내용 확인.
```

```python
model.visualize_barchart(top_n_topics=8) # 높은 비중을 차지하는 8개의 topic의 내용 시각화(barchart)
```

```python
# 주제간 거리 지도
model.visualize_topics()
```

```python
model.visualize_hierarchy(top_n_topics=10)
```

```python
model.visualize_heatmap(top_n_topics=10)
```

```python
# 특정 문서의 주제 분포 시각화
model.visualize_distribution(model.probabilities_[0], min_probability=0.015)
```

```python
########## 임베딩 벡터
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

```

```python
# BERTopic 모델 설정
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
topic_model = BERTopic(embedding_model=embedding_model)

topics, probabilities = topic_model.fit_transform(docs)

# 임베딩 모델을 사용하여 직접 임베딩을 생성
# 즉 인코더 최종 출력 임베딩 결과
embeddings = embedding_model.encode(docs)
print("임베딩 차원:", embeddings.shape)
print("첫 번째 문서의 임베딩 벡터:", embeddings[0])
```
