### 주성분 분석(차원 축소)

- 눈으로 보이게끔 차원을 줄이는 과정
- 피쳐가 많을수록 과적합 발생확률이 높아진다.(차원의 저주)
- 데이터 특성을 잘 표현하는 애들을 찾아서 그것만 표현하자

**주성분 분석 전 스케일링은 필수!**

```python
from sklearn.datasets import load_breast_cancer, load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA # 차원 축소

import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = pd.read_csv("../../data/heart.csv")
model_std = StandardScaler()
data_std = model_std.fit_transform(data[data.columns[:-1]])

data_pca = PCA(n_components = 2).fit_transform(data_std)

colors = {0:"blue", 1:"red"}
c = data["output"].replace(colors)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(data_pca[:,0], data_pca[:,1], alpha=0.6, color=c)
ax.set(xlabel=R"X", ylabel=R"Y", title="PCA");
```
