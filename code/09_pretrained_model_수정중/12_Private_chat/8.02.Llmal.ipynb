```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# LLaMA 모델 and 토크나이저 불러오기
# 'decapoda-research/llama-7b-hf': LLaMA 7B 모델의 Hugging Face 경로.
tokenizer = AutoTokenizer.from_pretrained("baffo32/decapoda-research-llama-7B-hf", use_fast=False)  # Disable fast tokenizer
model = AutoModelForCausalLM.from_pretrained("baffo32/decapoda-research-llama-7B-hf")
```

```python
# 텍스트 생성 위한 프롬프트
prompt = "인공지능이란 무엇인가요?"
```

```python
# 사용 가능한 장치 확인 (GPU 또는 CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# 입력된 프롬프트 토크나이징
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(device)
```

```python
# 텍스트 생성을 위해 모델에 입력 전달
# max_length : 생성할 텍스트 최대 길이
output = model.generate(input_ids, max_length=100, num_return_sequences=1)
```

```python
# 생성된 텍스트를 디코딩, 확인 가능한 형식으로 변환
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
# 결과 출력
print("입력된 프롬프트:", prompt)
print("생성된 텍스트:", generated_text)
```
