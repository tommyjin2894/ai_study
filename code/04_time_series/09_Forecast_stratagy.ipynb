### 예측 모델 정의하기
- 예측  시점에 이용 가능한 (피쳐)
- 예측 값이 필요한 기간 (타겟)

### 예측 기원 (forecast origin)
- 예측이 시작된 지점, 즉 훈련 데이터의 마지막 지점.
- 이 지점 까지 features 를 만들 수 있다.

### **forecast horizon**
보통 예측의 타겟 값으로, 스텝으로 표현한다.
origin 으로 부터 horizen 까지 떨어진 시간 : lead time(latency)

### 용어정리
| **용어** | **설명** |
| --- | --- |
| **피쳐 (Features)** | 예측 기원 시점까지의 데이터를 기반으로 생성. |
| **타겟 (Target)** | 예측 기간에 해당하는 미래의 값을 의미합니다. |
| **예측 기원 (Forecast Origin)** | 예측이 시작되는 시점, 훈련 데이터의 마지막 시점 <br>이 시점까지의 데이터로 피쳐를 생성 |
| **예측 기간 (Forecast Horizon)** | 예측이 필요한 기간, 스텝으로 표현 <br> 예측 기원으로부터 얼마나 떨어진 미래를 예측할 것인지 나타낸다. |
| **리드 타임 (Lead Time)** | 예측 기원과 예측 기간 사이의 간격 <br> 예측 기원으로부터 예측하고자 하는 미래 시점까지의 시간적 거리.  |

### 멀티 스텝 예측 전략!
- 멀티 아웃풋 (Multioutput Model)
    - 모델을 사용해서 한번에 여러 예측을 한번에 한다.
    - 선형회귀, NN 은 멀티 아웃풋을 낼 수 있다.
    - XGBoost는 할 수 없다.
- 직접 전략! (Direct Strategy)
    - 하나의 모델로 1단계 예측을 수행하고, 그 다음 예측 … 하나씩 수행한다.
- 재귀 전략? (**Recursive strategy)**
    - 하나의 모델로 1단계 예측하고 다음 지연 피쳐를 업데이트한다. 단계별로 오차가 전염되어 긴 예측에 점점 정확도가 떨어진다.

- 직접 + 재귀 전략 (**DirRec strategy)**
    - 모델을 학습하고 예측한다.
    - 예측 값을 피쳐로 생성해 다음 스텝을 예측
    - 반복
