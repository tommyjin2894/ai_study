```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("../stussy.png")

def ims(img):
    cv2.imshow('image',img)

    cv2.waitKey(1)
    while True:
        if cv2.waitKey() == ord("d"):
            break
    cv2.destroyAllWindows()
```

```python
img_to_BRT = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)


ims(img_to_BRT)
plt.imshow(img_to_BRT);
```

```python
rgb_img_copied = img_to_BRT.copy()
cv2.line(rgb_img_copied, (200,50), (300,100), (255,0,0), 2)
cv2.putText(rgb_img_copied, 'this is stussy',(80, 50), cv2.FONT_HERSHEY_PLAIN, 1, (255,0,0), 1,cv2.LINE_AA)
plt.imshow(rgb_img_copied);
```
