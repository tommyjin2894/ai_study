## Isotropic Multivariable Normal Distribution(등방성 다변수 정규 분포)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
mean = np.array([0, 0])  # Mean vector
variance = 1.0           # Common variance
cov_matrix = variance * np.identity(2)  # Covariance matrix for a 2D isotropic distribution

# Generate random samples from the isotropic multivariate normal distribution
num_samples = 1000
samples = np.random.multivariate_normal(mean, cov_matrix, num_samples)

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], marker='.', alpha=0.5)
plt.title('Isotropic Multivariate Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal')  # Make the aspect ratio of the plot equal
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
mean = np.array([0, 0, 0])  # Mean vector
variance = 1.0             # Common variance
cov_matrix = variance * np.identity(3)  # Covariance matrix for a 3D isotropic distribution

# Generate random samples from the isotropic multivariate normal distribution
num_samples = 1000
samples = np.random.multivariate_normal(mean, cov_matrix, num_samples)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(samples[:, 0], samples[:, 1], samples[:, 2], marker='.', alpha=0.5)

# Set plot labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Isotropic Multivariate Normal Distribution in 3D')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
mean = np.array([0, 0, 0])  # Mean vector
variance = 1.0             # Common variance
cov_matrix = variance * np.identity(3)  # Covariance matrix for a 3D isotropic distribution

# Create a 3D grid of points
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)
z = np.zeros_like(x)

# Calculate the PDF values for each point in the grid
for i in range(len(x)):
    for j in range(len(y)):
        point = np.array([x[i, j], y[i, j], 0])
        z[i, j] = (1 / (2 * np.pi * np.sqrt(np.linalg.det(cov_matrix)))) * \
                  np.exp(-0.5 * (point - mean).T @ np.linalg.inv(cov_matrix) @ (point - mean))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set plot labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('PDF')
ax.set_title('Isotropic Multivariate Normal Distribution in 3D')

plt.show()
```
