### 인공지능의 윤리적 고려사항

- 도덕적이고 책임감 있게 사용해야한다.
윤리의 5대 문제
  - 편향성 : 데이터의 편향적 으로 인한 사회적 불평등
  - 악용 : 딥페이크, 피싱, 사이버 범죄 및 불법적 이용
  - 개인정보 보호 : 수집하고 분석시 오는 사생활 침해
  - 킬러 로봇 : 실제 전쟁에 사용한다.
  - 오류와 안정성 : 자율사고 문제 및 의료 진단 문제등

필요성
  - 특성을 감안한 필요성 : 자율성과 지능의 특성 - 점점 발전한다. 인간의 제어를 뛰어 넘음.
  - 발생 사건에 근거한 필요성 : 누가 책임을 질 것인가.(차량 사고, 및 주식)
  - 발생 가능을 상상을 근거한 필요성 : 미래의 문제에 대하여 미리 대비 한다.
  - 인터넷 윤리의 연장선으로서의 필요성 : 개인정보 유출이나 , 빈익빈 부익부 현상
  - 경쟁력 확보 : 글로벌 표준화, 경쟁력 영향, 윤리적 대비, 글로벌 표준화와의 경쟁력
기준
  - 현재 가이드라인이 있다 : 기술의 순기능을 최대화, 역기능 최소화
  - 범 국가 차원의 기본 원칙 : 
    - 존엄성 : 정신적 신체적 건강에 해가 되지 않는 범위 
    - 공공성 : 다수의 사람을 위한 행복추구 및 접근성 원칙, 인류의 보편적 복지를 향상시켜야함
    - 기술의 합목적성 : 인류의 삶에 필요한 도구의 목적 의도에 부합 되게 하여야함
    등등

윤리는 자율규범이다 : 현제 아직 파악중이므로 기술 발전과 산업의 성장을 저해하지 않는 한에서 법을 제정하고 있는중이다.

개발자의 역할 :
  - 이해 관계자의 의견을 수렴해야한다.
  - 명확한 윤리적 기준을 세워야한다.
  - 지속적인 교육과 훈련
  - 국제적 협력과 표준화 작업
