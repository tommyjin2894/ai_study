### 주성분 분석 (Principal Component Analysis, PCA)
* PCA는 데이터를 시각적으로 이해하기 쉽도록 차원을 축소하고, 모델의 일반화를 돕는다.

1. **차원 축소 방법**:  
   PCA는 데이터의 중요한 특성만 추출해 차원을 줄이는 대표적인 차원 축소 기법입니다.

2. **과적합 방지**:  
   피처가 많아질수록 과적합 위험이 커지기 때문에, PCA로 중요한 피처만 남겨 모델 성능을 향상할 수 있습니다.

3. **주성분 찾기**:
   - **분산이 최대인 방향**을 찾아 해당 방향을 첫 번째 **주성분**으로 설정합니다.
   - 첫 번째 주성분에 **수직이면서 분산이 큰 방향**을 두 번째 주성분으로 설정하며, 이런 방식으로 데이터 특성 수만큼 주성분을 찾습니다.

4. **스케일링 필수**:  
   PCA 전에는 데이터 **스케일링**이 필수입니다.

```python
from sklearn.datasets import load_breast_cancer, load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA # 차원 축소

import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = pd.read_csv("../../data/heart.csv")
model_std = StandardScaler()
data_std = model_std.fit_transform(data[data.columns[:-1]])

data_pca = PCA(n_components = 2).fit_transform(data_std)

colors = {0:"blue", 1:"red"}
c = data["output"].replace(colors)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(data_pca[:,0], data_pca[:,1], alpha=0.6, color=c)
ax.set(xlabel=R"X", ylabel=R"Y", title="PCA");
```

### 데이터 스케일링

```python
## 데이터를 standard scaling 해보자
sta = StandardScaler()
cancer_scaling = sta.fit_transform(cancer_df[cancer_df.columns[:-1]])
```

### PCA

```python
cancer_pca = PCA(n_components = 2).fit_transform(cancer_scaling)
cancer_pca
```

```python
colors = {0 : "red", 1: "blue"}
c = cancer_df["label"].replace(colors)
```

```python
fig, ax= plt.subplots(figsize=(4,4))
ax.scatter(cancer_pca[:,0], cancer_pca[:,1], s = 15, alpha = 0.6, color=c)
ax.set(xlabel=R"$X$", ylabel=R"$Y$", title = "PCA");
```

### PCA

```python
## 데이터를 pca하여 시각화 해보자.

digits_pca = PCA(n_components = 2).fit_transform(digits["data"])

colors = ["C0", "C1", "C2", "C3", "C4", "C5", "C6", "C7", "C8", "C9"]
c = [colors[i] for i in digits["target"]]

fig, ax = plt.subplots(figsize=(4,4))
ax.scatter(digits_pca[:,0], digits_pca[:,1], s = 15, alpha=0.6, color = c)
ax.set(xlabel=R"$X$", ylabel=R"$Y$", title = "PCA");
```
