- ì¼í‘œë³¸ ê²€ì • (One sample test)
    - ì •ê·œì„± ê°€ì •ì„ ë§Œì¡±í•˜ë©´: ì¼í‘œë³¸ t ê²€ì • (One sample t-test)
    - ì •ê·œì„± ê°€ì •ì„ ë§Œì¡±ì•ˆí•¨: ìœŒì½•ìŠ¨ì˜ ë¶€í˜¸ ìˆœìœ„ ê²€ì • (wilcoxon's signed rank test)

```python
# 1. ê°€ì„¤ ê²€ì • ì²«ë²ˆì§¸
# ê·€ë¬´ ê°€ì„¤: ì„±ì¸ë“¤ì˜ ìš©ëˆì˜ í‰ê· ì€ 50ë§Œì›ì´ë‹¤.
# ëŒ€ë¦½ ê°€ì„¤: ì„±ì¸ë“¤ì˜ ìš©ëˆì˜ í‰ê· ì€ 50ë§Œì› ë³´ë‹¤ ë§ë‹¤.

# ì—‘ì…€ ë°ì´ë¼ ì½ì–´ì˜¤ê¸°
income = pd.read_csv("../../../data/income.csv", index_col=0)
income
```

```python
# 1ë‹¨ê³„ : ì •ê·œì„± ê²€ì •(Normality Test)
# ê·€ë¬´ ê°€ì„¤ : ì„±ì¸ë“¤ì˜ ìš©ëˆì€ ì •ê·œ ë¶„í¬ë¥¼ ë”°ë¥¸ë‹¤. H_0
# ëŒ€ë¦½ ê°€ì„¤ : ì„±ì¸ë“¤ì˜ ìš©ëˆì€ ì •ê·œ ë¶„í¬ë¥¼ ë”°ë¥´ì§€ ì•ŠëŠ”ë‹¤. H_1

# n(í‘œë³¸ì˜ ë°ì´í„° ìˆ˜) < 5000 : Shapiro-wilk
# n(í‘œë³¸ì˜ ë°ì´í„° ìˆ˜) >= 5000 : Anderson-darling

# n = 19 < 5000 : Shapiro-wilk
stats.shapiro(income["money"])
```

```python
# statistic=0.975 : ê²€ì • í†µê³„ë‘ W
# pvalue=0.866    : ìœ ì˜ í™•ë¥ 

# ê²°ë¡ 
# p-value ê°€ (0.866) > ìœ ì˜ìˆ˜ì¤€ 0.05 : ê·€ë¬´ê°€ì„¤
# ì„±ì¸ë“¤ì˜ ìš©ëˆì€ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥¸ë‹¤. ->>>>> ì›ìƒ˜í”Œ T í…ŒìŠ¤íŠ¸ë¥¼ í•œë‹¤.
```

```python
# 2ë‹¨ê³„ : ì¼í‘œë³¸ tê²€ì • (one sample t-test)
stats.ttest_1samp(income["money"],
                  popmean     = 50,
                  alternative = "greater")

# income["money"] : ê²€ì • ë³€ìˆ˜
# popmean         : ê·€ë¬´ê°€ì„¤ì˜ ëª¨í‰ê· 
# alternative     : ëŒ€ë¦½ê°€ì„¤ : "greater", "less", "two-sided."
```
