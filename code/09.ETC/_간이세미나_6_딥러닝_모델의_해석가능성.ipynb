# 해석

선형 회귀, 결정 트리 = 해석의 용이
딥러닝 = 설명 제공하지 않기 떄문에 알 수 없고, 복잡성에 따라 해석이 어려움

1 Model Agnostic : 모델의 관계없이 사용 가능
2 model Specific : 특정 모델에서만 사용 가능

국소적 해석 : 일부 예측에 대한 해석 - 대출 거절

전역적 해석 : 전반적인해석 - 전반적인 합리적인 기준을 판별하고 있는가

LIME 기법

Local Inter~

사후 분석 : 결과에 대한 해석
모델의 일부분을 보고 해석

해석 가능한 모델을 근사시켜 예측을 근사 (Linear model 이나 DT)

되도록 이분법적인 피쳐로 변환해서 트레인 한다.



SHAP 모델 <br>

Shaply Value : 게임 이론 <br>
게임에서 각 플레이어가 얼마나 기여했는지 확인 <br>

각 피쳐별로 얼마나 기여했는지 확인 <br>

각각 국소적 기여도를 계산할수 있고 <br>
모든 shapy valu 를 구해 전역적으로도 해석이 가능하다 <br>



CAM
Class Activation Map
CNN 에서 사용 가능하다
CNN 마지막 레이어를 변경

마지막 레어를 GAP 레이어로 교체하는 방식으로 사용 <br>
필터 전체를 대표값으로 통합한다.(평균) <br>
필터의 가중합을 구해 클래스별로 피쳐맵을 만들어 캠의 각 값은 클래스 판별에 얼마나 중요한지  <br>
추후 예측을 할때 업스케일 하여 중요도를 파악 할 수 있다. <br>

Grad cam 마지막 레이어를 결과를 변경하기 대신
피쳐의 각노드별 확률에대한 기울기를 구하여 평규을 냈을때 가중치와 비슷한 역할을 하게된다 (모델의 구조를 바꾸지 않고도 구할 수있다.)

좋은 해석

왜? 라는 질문에 대한 타당한 대답이다

설명에는 결국 사람의 주관이 들어갈 수 밖에 없다.

기준
1. 대조적 설명 : 비교
2. 핵심 요인 선택 : 모든 일에는 다양한 원인이 있지만, 그 중에 핵심 원인
3. 사회적 : 설명은 대상과 환경에 따라 달라져야한다.
