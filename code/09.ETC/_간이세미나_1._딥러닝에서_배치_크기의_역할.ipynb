# 배치 사이즈란?

- Epoch : 전체 데이터 셋에 대하여 전체 학습의 완료 주기
- Batch : 한번의 학습 (Iterartion) 에 대한 데이터의 양
- Iteration : 배치를 처리하는 단위

즉 배치 사이즈 마다 가중치가 업데이트 된다.

오차의 **평균**이 작아지는 방향으로 가중치가 업데이트된다.

## 배치 사이즈별 학습

- 배치 사이즈가 클 때
    - 병렬 연산에 효율적
    - 메모리 사용량이 큼
    - 경사도가 평준화 되어 노이즈가 적다.
- 배치 사이즈가 작을 때
    - 가중치가 너무 자주 업데이트 되어 시간이 오래 걸린다.
    - 진동이 커진다. 경로를 너무 업데이트 되어 노이즈가 많이 끼기 때문
    - 진동이 커져 학습이 불안정 할 수 있다.
    - 다양한 경로를 탐색하여 일반화 성능이좋다.

## 결론:

배치 사이즈 마다 가중치 업데이트

배치 사이즈 크기마다 오차 값이 매우 달라진다.

트레이드 오프 (최적 점) 을 잘 찾아야 한다.

## 배치 사이즈 선택 요령

1. 2의 배수 로 선정, 보통 32, 64, 128, 256 등의 크기로 배치사이즈를 정한다.
2. 학습 률이 크면 배치 사이즈도 크게
3. 배치 사이즈의 변화량은 학습률의 변화량과 비례적으로 조정한다.
    
    $$
    \mu_{new} = \mu_{old} \times \frac{\text{batch size}_{new}}{\text{batch size}_{old}}
    $$
