# 활성화 함수

1. 의미 역할
2. 종류
3. 선택 기준

# 의미와 역할
- 논 리니어 형태로 변환

# 종류
- 계단 함수
- 시그모이드
  - 논 제로 센터드 프로블럼 : 모든 수가 양수인 문제 - 가중치의 값이 항상 + 이기떄문에 효율이 안좋아 진다 지그재그. 은닉은 잘 사용하지 않는다. 출력에서 잘 사용한다.
- 하이퍼 볼릭 탄젠트 함수 : Nonzero를 보완한다. 여전히 기울기 소실이 있다. 상대적 빠른
- 렐루 : 빠른 훈련 속도, 대부분 은닉층에서 사용, 0 보다 작은 값의 정보가 사라진다.
  - 리키렐루, PReLU, EReLU
- 소프트 맥스 : 확률적 개념 모든 총합은 1 상대평가 가능, 다중 분류 가능

# 선택 기준
1. 문제의 특성 : 분류 회귀
2. 신경망 깊이 : 기울기 소멸 문제
3. 계산 문제 : 활성화 함수의 계산 복잡도~
